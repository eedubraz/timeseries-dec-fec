### Simple Exponential Smoothing

*A value close to 1 indicates fast learning (that is, only the most recent values influence the forecasts), whereas a value close to 0 indicates slow learning (past observations have a large influence on forecasts).*

```python
y_hat = df_test.copy()
fit2 = SimpleExpSmoothing(np.asarray(df_train['CHI_COM_EXP'])).fit(smoothing_level = 0.6,optimized = False)
y_hat['SES'] = fit2.forecast(len(df_test))
plt.figure(figsize =(15,8))
plt.plot(df_train['CHI_COM_EXP'], label = 'Treino')
plt.plot(df_test['CHI_COM_EXP'], label = 'Validação')
plt.plot(y_hat['SES'], label = 'Simple Exponential Smoothing', color='b')
plt.legend(loc = 'best')
```

```python
rmse_SES = sqrt(mean_squared_error(df_test['CHI_COM_EXP'], y_hat['SES']))
rmse_SES
```

```python
def SES_forecast(x):
    y_hat = df_test.copy()
    fit2 = SimpleExpSmoothing(np.asarray(df_train['CHI_COM_EXP'])).fit(smoothing_level = x,optimized = False)
    y_hat['SES'] = fit2.forecast(len(df_test))
    rmse_SES = sqrt(mean_squared_error(df_test['CHI_COM_EXP'], y_hat['SES']))
    print(round(rmse_SES,2))
```

```python
for i in np.arange(0.0,1.0,0.01):
    print('Valor smoothing: ' + str(round(i,2)))
    SES_forecast(i)
```

```python
SES_forecast(0.01)
```

### Holt's Linear Trend Model

```python
plt.style.use('ggplot')
plt.figure(figsize = (16,8))
import statsmodels.api as sm
sm.tsa.seasonal_decompose(df_train['CHI_COM_EXP']).plot()
result = sm.tsa.stattools.adfuller(df_train['CHI_COM_EXP'])
plt.show()
```

```python
y_hat_holt = df_test.copy()
fit1 = Holt(np.asarray(df_train['CHI_COM_EXP'])).fit(smoothing_level = 0.1, smoothing_slope = 0.1)
y_hat_holt['Holt_linear'] = fit1.forecast(len(df_test))
plt.style.use('fivethirtyeight')
plt.figure(figsize = (15,8))
plt.plot(df_train['CHI_COM_EXP'], label = 'Train')
plt.plot(df_test['CHI_COM_EXP'], label = 'Validation')
plt.plot(y_hat_holt['Holt_linear'], label = 'Holt Linear')
plt.legend(loc = 'best')
```

```python
rmse_holt = sqrt(mean_squared_error(df_test['CHI_COM_EXP'], y_hat_holt.Holt_linear))
rmse_holt
```

### Holt Winter's Model

```python
y_hat_avg = df_test.copy()
fit1 = ExponentialSmoothing(np.asarray(df_train['CHI_COM_EXP']), seasonal_periods= 365, seasonal= 'mul').fit()
y_hat_avg['Holt_Winter'] = fit1.forecast(len(df_test))
plt.figure(figsize = (16,8))
plt.plot(df_train['CHI_COM_EXP'], label = 'Train')
plt.plot(df_test['CHI_COM_EXP'], label = 'Validation')
plt.plot(y_hat_avg.Holt_Winter, label = 'Holt Winters')
plt.legend(loc = 'best')
```

```python
rmse_holt_winters = sqrt(mean_squared_error(df_test['CHI_COM_EXP'], y_hat_avg['Holt_Winter']))
rmse_holt_winters
```
